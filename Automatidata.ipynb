```python
# Convert data columns to datetime
df['tpep_pickup_datetime']= pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime']= pd.to_datetime(df['tpep_dropoff_datetime'])
```

```python
# Create box plot of trip_distance
plt.figure(figsize=(10,3))
plt.title('trip_distance')
sns.boxplot(data=None, x=df['trip_distance'], fliersize=1)
```

```python
# Create histogram of trip_distance
plt.hist(df['trip_distance'], bins= range(0,30,1), edgecolor= "black")
plt.title("Trip Distance Histogram")
plt.show()
```

```python
#Create Histogram of Tip_amount by vendor
sns.histplot(
    data=df,
    x='tip_amount',
    bins=range(0, 21, 1),
    hue='VendorID',
    multiple='stack',
    palette='pastel'
)

plt.xticks(range(0, 21, 1))
plt.show()
```

```python
tip_amount_1= df[df['tip_amount'] >10]
sns.histplot(
    data=tip_amount_1,
    x='tip_amount',
    bins=range(10, 21, 1),
    hue='VendorID',
    multiple='stack',
    palette='pastel'
)

plt.xticks(range(10, 21, 1))
plt.show()
```

```python
#Examine the unique values in the passenger_count column
df['passenger_count'].value_counts()
```

```python
df['tip_amount']
```

```python
# Calculate mean tips by passenger_count
mean_tips_by_passenger_count = (
    df.groupby('passenger_count', as_index=False)['tip_amount']
      .mean()
      .rename(columns={'tip_amount': 'mean_tip_amount'})
)
mean_tips_by_passenger_count
```

```python
#create  a bar plot for mean tips by passenger count
plt.bar(df["passenger_count"], df["mean_tip_amount"])
plt.xlabel("Passenger Count")
plt.title("Mean Tips by Passenger Count")
plt.show()
```
